```python
#ml
import pandas as pd 
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt 

# ml library 
from sklearn import *  
print('done')
```

```python
# prob 
data ={ 'Study_hrs': range(1,11),
       'Marks Obtained':[25,33,46,50,55,67,78,85,95,100]}
df = pd.DataFrame(data)
df
```

```python
# ml model build - scatter plot , box plot 
```

```python

plt.title('Study vs Marks')
plt.scatter(df['Study_hrs'],df['Marks Obtained'],marker = 'x', color = 'r')
plt.xlabel('Study Hrs')
plt.ylabel('Marks')
plt.show()
```

```python
# x(1) , y(1) : simple linear regression 
# x( multiple) , y (1) : multiple linear regression 

```

```python
# import module 
from sklearn.linear_model import LinearRegression # class
# object model creation 
model = LinearRegression() 
# x , y part 
X = df[['Study_hrs']] # due to data frame and take capital of x to pass multiple value 
y = df['Marks Obtained']

# trainning  / learning 
model.fit(X,y)
```

```python
X =[[5,5]]

m = model.coef_
c = model.intercept_

print(m,c)
```

```python
#y = m*X +c
#y = w*X +c

std_hrs=[[5.5]]
marks = m *std_hrs +c 
print('predicted marks:',marks[0][0])
```

```python
import warnings 
warnings.filterwarnings('ignore')

study_hrs = input(" enter the no of study hours")
study_hr = [[float(study_hrs)]]

ans = model.predict(study_hr)
marks = round(ans[0],2)

if marks>100:
    marks =100
elif marks<0:
    marks =0
else:
    marks = marks
    
print(f'your predicted  marks is : {marks}')
```
