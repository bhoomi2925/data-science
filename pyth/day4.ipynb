```python
from googlesearch import search
title ='free python pdf'

data = search(title,num_results=10)

all_data =[]

for i in data:
    all_data.append(i + '\n')

file = open(f'{title}.txt', mode = 'w')
file.writelines(all_data)
file.close()

print('done')
    
    
```

```python
from googlesearch import search
title ='chat gpt trending news and investement '

data = search(title,num_results=5)

all_data =[]

for i in data:
    all_data.append(i + '\n')

# it will append new line :a 
file = open(f'{title}.txt', mode = 'a')
file.write('===============\n')
file.writelines(all_data)
file.close()

print('done')
```

```python
url = '''https://picsum.photos/1080/720'''
import requests as r 
n=10
for i in range(n):
    img_data = r.get(url).content
    f = open (f'image_{i}.jpg','wb')
    f.write(img_data)
    f.close()
    from IPython import display as d 
    display(d.Image(f'image_{i}.jpg',width = 200,height = 200))
```
