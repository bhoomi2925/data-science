```python
# Internet Required
df_titanic = sns.load_dataset('titanic')
df_titanic.sample()
```

```python
import matplotlib.pyplot as plt
sns.countplot(data = df_titanic, x = 'sex', hue = 'alive',palette = sns.color_palette('magma',2)) # hue: further divide: based on categorical(object)
plt.title('Titanic survival analysis based on gender')
plt.show()
```

```python
# sns.barplot 
sns.barplot(data = df_titanic , x = 'class' , y = 'fare', hue ='alive', errorbar=('ci',0))  # error bar denotes as ci 
plt.grid(color ='g' , alpha = 0.2)
plt.show()
# black line denotes confidence interval(CI) = 95% , estimator i to check aggeregation 
```

```python
#sns.boxplot : to detect outliers to check data distribution 
# to detect outliers : 
#range: min to max (19-22) : 35 , 28 ,45(outliers :fliers) it affect on mean 
# Q2: Quantile : 50%
# Q1: 25%
# Q3: 75%

# IQR = Q3 - Q1 ( inter quartile range )
#UW = Q3 + 1.5*IQR ( upper cap) 
#LW = Q1 - 1.5* IQR ( lowe cap)
```

```python
sns.boxplot(data = df_titanic,x ='class',y='fare',hue ='sex')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data = df_titanic,y ='class',x='fare',hue ='sex',orient='h') # outlier , secrewness , median value check 
plt.show()
```

```python
#sns.violin.plot
plt.figure(figsize=(10,5))
sns.violinplot(data = df_titanic,x ='class',y='fare',hue ='sex',orient='v') # outlier , secrewness , median value check 
plt.show()
```

```python
#sns.pairplot(num vs num)
df_iris = sns.load_dataset('iris')
df_iris.sample()
```

```python
sns.pairplot(data=df_iris , hue ='species',corner = True)
plt.show()
```

```python
# sns.kdeplot : kernel density estimation 
sns.kdeplot(data = df_iris, x ='sepal_length' ,y = 'sepal_width',hue = 'species')
# sns.scatterplot(data = df_iris, x ='sepal_length' ,y = 'sepal_width',hue = 'species')
plt.show()
```

```python
# jointplot ( num vs num)
sns.jointplot(data = df_iris, x ='sepal_length' ,y = 'sepal_width',hue = 'species')
plt.show()
```

```python
sns.jointplot(data = df_iris, x ='sepal_length' ,y = 'sepal_width',kind = 'reg')
plt.show()
```
